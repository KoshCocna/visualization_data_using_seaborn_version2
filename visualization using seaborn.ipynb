```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips['tip_percentage'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.histplot(tips['tip_percentage'], binwidth=0.05)
```

```python
sns.histplot(data=tips, x='tip_percentage', hue='time', binwidth=0.05)
```

```python
import numpy as np
sns.barplot(data=tips, x='day', y='tip', estimator=np.sum)
```

###     

```python
plt.figure(figsize = (5,3))
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = "Set3")
plt.show()
```

###   lmplot 

```python
sns.set_style("darkgrid")
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, height = 4)
plt.show()
```

hue  smoker    

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker', palette = 'Set2')
plt.show()
```

###    flights  

```python
flights = sns.load_dataset('flights')
flights.head()
```

pivot     

```python
flights = flights.pivot(index='month', columns='year', values='passengers')
flights
```

```python
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```

```python
flights = sns.load_dataset('flights')
sns.scatterplot(data=flights, x = 'year', y='passengers')
```
